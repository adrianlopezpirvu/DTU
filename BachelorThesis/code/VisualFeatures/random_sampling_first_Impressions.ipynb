```python
import pandas as pd
import os
import zipfile
```

```python
gender_annotations_df = pd.read_csv('../FirstImpression/eth_gender_annotations_dev.csv', delimiter=';')
age_annotations_df = pd.read_csv('../FirstImpression/fi_age_labels/age_anno_dev.csv', delimiter=',')

# Merge
merged_df = pd.merge(gender_annotations_df, age_annotations_df, on='VideoName')

merged_df
```

```python
# Define el nÃºmero de muestras por categorÃ­a
num_samples_per_category = 1

# Define las categorÃ­as
ethnicities = [1, 2, 3]
genders = [1, 2]
age_groups = [4, 5, 6]

sampled_rows = []

# Muestrea aleatoriamente de cada categorÃ­a
for ethnicity in ethnicities:
    for gender in genders:
        for age_group in age_groups:
            # Filtra el DataFrame por las categorÃ­as actuales
            filtered_merged_df = merged_df[(merged_df['Ethnicity'] == ethnicity) & 
                             (merged_df['Gender'] == gender) & 
                             (merged_df['AgeGroup'] == age_group)]
            # Si hay suficientes muestras en la categorÃ­a actual, selecciona aleatoriamente
            if len(filtered_merged_df) >= num_samples_per_category:
                sampled_rows.extend(filtered_merged_df.sample(n=num_samples_per_category).index.tolist())

# Crea un nuevo DataFrame con las filas muestreadas
sampled_merged_df = merged_df.loc[sampled_rows]


sampled_merged_df
```
